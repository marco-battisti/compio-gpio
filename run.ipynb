```python
import matplotlib.pyplot as plt
import re as regex
import pandas

PANDAS_DATA = None
file_path = "data.csv"

with open(file_path, "r") as f:
    temp = pandas.read_csv(file_path)
    PANDAS_DATA = temp.copy()

    last_col_list = PANDAS_DATA.columns[-1]
    PANDAS_DATA = PANDAS_DATA[PANDAS_DATA.columns[:-1]]
    col_data = {}

    for index, column_name in enumerate(regex.split(r'[\t]+', last_col_list)):
        if not column_name: continue

        line = temp.iloc[[index]]
        line = line[line.columns[-1]]

        col_data[column_name] = regex.split(r'[\t]+', line.values[0])[index]

    for column_name, column_data in col_data.items():
        PANDAS_DATA[column_name] = column_data


#print(PANDAS_DATA)

F_infortunio_rate_non_fatal = PANDAS_DATA.query('sex == "F" & age == "TOTAL" & unit == "PC" & severity == "D_GE4"')

print(F_infortunio_rate_non_fatal)

#plt.plot(
#    
#)
#plt.plot(
#    PANDAS_DATA.query('sex == "M" & age == "TOTAL"').get("2008")
#)
#plt.show()
```
